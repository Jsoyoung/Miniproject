```python
def calculate_unit_price(data):
    result = {}
    
    for index,row in data.iterrows():
        house_type = row['HOUSE_TYPE']  # 건물 용도
        tot_area = float(row['TOT_AREA'])  # 토지 면적(제곱미터)
        transaction_price = float(row['OBJ_AMT'])  # 물건 금액(만원)
        
        if house_type not in result:
            result[house_type] = {'total_area': 0, 'total_price': 0, 'count': 0}
            
        result[house_type]['total_area'] += tot_area  # 건물 용도별 토지면적 누적
        result[house_type]['total_price'] += transaction_price  # 건물 용도별 물건 금액 누적
        result[house_type]['count'] += 1  # 건물 용도별 거래 건수 누적
        
    for house_type, values in result.items():
        if values['total_area'] != 0:
            average_price_per_tot_area = values['total_price'] / values['total_area']
        else:
            average_price_per_tot_area = 0
            
        result[house_type]['average_price_per_tot_area'] = average_price_per_tot_area
    
    return result
    
result = calculate_unit_price(data)
    
for house_type, values in result.items():
    print(f'{house_type}: {values["average_price_per_tot_area"]:.2f} 원/㎡')
```

```python
import pandas as pd


# 각 건물용도별 최대값 초기화
max_values = {
    '아파트': {'amount': None, 'location': None, 'building': None},
    '오피스텔': {'amount': None, 'location': None, 'building': None},
    '연립다세대': {'amount': None, 'location': None, 'building': None},
    '단독다가구': {'amount': None, 'location': None, 'building': None}
}

for house_type in ['아파트', '오피스텔', '연립다세대','단독다가구']:
    # 해당 건물용도의 데이터 필터링
    filtered_data = df[df['HOUSE_TYPE'] == house_type]
    
    # 거래금액('OBJ_AMT') 열에서 최대 값 찾기
    max_amount = filtered_data['OBJ_AMT'].max()
    
    # 최대값을 가진 행 찾기
    max_row = filtered_data[filtered_data['OBJ_AMT'] == max_amount].iloc[0]

    # 최대값 및 위치 정보 저장
    max_values[house_type]['amount'] = max_amount
    max_values[house_type]['location'] = max_row['SGG_NM'] if not pd.isna(max_row['SGG_NM']) else ''
    max_values[house_type]['location'] += ' ' + max_row['BJDONG_NM'] if not pd.isna(max_row['BJDONG_NM']) else ''
    max_values[house_type]['building'] = max_row['BLDG_NM'] if not pd.isna(max_row['BLDG_NM']) else ''

# 결과 출력
for house_type, values in max_values.items():
    print(f"{house_type} 최대값:", values['amount'])
    print("자치구 및 법정동:", values['location'], values['building'])
```

```python


# 각 건물용도별 최대값 및 최소값 초기화
values = {
    '아파트': {'max_amount': None, 'min_amount': None, 'max_location': None, 'min_location': None, 'max_building': None, 'min_building': None},
    '오피스텔': {'max_amount': None, 'min_amount': None, 'max_location': None, 'min_location': None, 'max_building': None, 'min_building': None},
    '연립다세대': {'max_amount': None, 'min_amount': None, 'max_location': None, 'min_location': None, 'max_building': None, 'min_building': None},
    '단독다가구': {'max_amount': None, 'min_amount': None, 'max_location': None, 'min_location': None, 'max_building': None, 'min_building': None}
}

for house_type in ['아파트', '오피스텔', '연립다세대','단독다가구']:
    # 해당 건물용도의 데이터 필터링
    filtered_data = df[df['HOUSE_TYPE'] == house_type]
    
    # 거래금액('OBJ_AMT') 열에서 최대 값 및 최소 값 찾기
    max_amount = filtered_data['OBJ_AMT'].max()
    min_amount = filtered_data['OBJ_AMT'].min()
    
    # 최대값 및 최소값을 가진 행 찾기
    max_row = filtered_data[filtered_data['OBJ_AMT'] == max_amount].iloc[0]
    min_row = filtered_data[filtered_data['OBJ_AMT'] == min_amount].iloc[0]

    # 최대값 및 최소값 및 위치 정보 저장
    values[house_type]['max_amount'] = max_amount
    values[house_type]['min_amount'] = min_amount
    values[house_type]['max_location'] = max_row['SGG_NM'] if not pd.isna(max_row['SGG_NM']) else ''
    values[house_type]['max_location'] += ' ' + max_row['BJDONG_NM'] if not pd.isna(max_row['BJDONG_NM']) else ''
    values[house_type]['max_building'] = max_row['BLDG_NM'] if not pd.isna(max_row['BLDG_NM']) else ''
    values[house_type]['min_location'] = min_row['SGG_NM'] if not pd.isna(min_row['SGG_NM']) else ''
    values[house_type]['min_location'] += ' ' + min_row['BJDONG_NM'] if not pd.isna(min_row['BJDONG_NM']) else ''
    values[house_type]['min_building'] = min_row['BLDG_NM'] if not pd.isna(min_row['BLDG_NM']) else ''

# 결과 출력
for house_type, data in values.items():
    print(f"{house_type} 최대값:", data['max_amount'])
    print("최대값 위치 정보:")
    print("  자치구 및 법정동:", data['max_location'],data['max_building'])
    print(f"{house_type} 최소값:", data['min_amount'])
    print("최소값 위치 정보:")
    print("  자치구 및 법정동:", data['min_location'], data['min_building'])
```
